```python
# Step 1: Split your data into two separate datasets for minority and majority class
X_minority = df[df['Bankrupt'] == 1]  # Your minority class data
X_majority = df[df['Bankrupt'] == 0]  # Your majority class data

# Step 2: Determine the desired number of observations for each class
length_minority = len(X_minority)
length_majority = len(X_majority)


oversampled_minority = resample(X_minority, random_state=5, n_samples = length_majority, replace = True)
#concatenate this with X_majority

undersampled_majority = resample(X_majority, random_state=5, n_samples = length_minority, replace = True)
#concatenate this with X_minority
```

```python
UDS = pd.concat([undersampled_majority, X_minority], ignore_index=True) #create undersampled dataset
```

```python
UDS_X = UDS.drop(columns=['Bankrupt']) #select all features except Y for undersampled dataset
UDS_y = UDS[['Bankrupt']] #Select Y for undersampled dataset
```

```python
UDS_X_train, UDS_X_test, UDS_y_train, UDS_y_test = train_test_split(UDS_X, UDS_y, test_size = 0.3, random_state = 4)
#create training and test sets on undersampled dataset
```

```python
df_X = df.drop(columns=['Bankrupt']) #select all features except Y
df_y = df[['Bankrupt']] #Select Y
```

```python
df_X.head() #check x
```

```python
df_y.head() #check y
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size = 0.3, random_state = 4)
#create training and test sets on original dataset
```

```python
#create a Random Forest Classifier on original dataset (model 1)
df_RFC = RandomForestClassifier(random_state=40)
```

```python
#fit the model on the training data (model 1)

%timeit -r 1 -n 1 df_RFC.fit(X_train, y_train) #3.32 seconds, model 1 takes longer to fit
```

```python
#make predictions on the test set
y_pred_df_RFC = df_RFC.predict(X_test)
```

```python
#Since classes are imbalanced, using ROC Curve to evaluate model instead of accuracy (model 1)
#Get predicted prob for positive class
df_RFC_y_prob = df_RFC.predict_proba(X_test)[:,1]

#Calculate ROC Curve
df_RFCfpr, df_RFCtpr, df_RFCthreshold = roc_curve(y_test, df_RFC_y_prob)

#Calculate the Area Under the Curve
df_RFCroc_auc = roc_auc_score(y_test, df_RFC_y_prob)

#Plot the ROC curve
plt.figure()
plt.plot(df_RFCfpr, df_RFCtpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % df_RFCroc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
#ROC Curve area = 0.94
#Model is 44% better than a coin-flip
```

```python
#create a Random Forest Classifier with different parameters on original dataset, only changing criterion to help answer question 4 accurately
df_RFC2 = DecisionTreeClassifier(criterion = 'entropy', random_state=40) # (model 2)
```

```python
#fit the model on the training data (model 2)

%timeit -r 1 -n 1 df_RFC2.fit(X_train, y_train) #286 milliseconds, model 1 takes longer to fit
```

```python
#make predictions on the test set
y_pred_df_RFC2 = df_RFC2.predict(X_test)
```

```python
#Since classes are imbalanced, using ROC Curve to evaluate model instead of accuracy (model 2)
#Get predicted prob for positive class
df_RFC2_y_prob = df_RFC2.predict_proba(X_test)[:,1]

#Calculate ROC Curve
df_RFC2fpr, df_RFC2tpr, df_RFC2threshold = roc_curve(y_test, df_RFC2_y_prob)

#Calculate the Area Under the Curve
df_RFC2roc_auc = roc_auc_score(y_test, df_RFC2_y_prob)

#Plot the ROC curve
plt.figure()
plt.plot(df_RFC2fpr, df_RFC2tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % df_RFC2roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
#ROC Curve area = 0.59
#Model is 9% better than a coin-flip
```
